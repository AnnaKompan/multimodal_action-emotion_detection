```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
```

```python
train_data = "./dataset/train"
test_data = "./dataset/test"
```

```python
classes = [d for d in os.listdir(train_data) if d != '.DS_Store']
print("Classes: ", classes)
```

```python
train_counts = {}
for cls in classes:
    cls_folder = os.path.join(train_data, cls)
    img = [im for im in os.listdir(cls_folder)]
    train_counts[cls] = len(img)
train_counts
```

```python
test_counts = {}
for cls in classes:
    cls_folder = os.path.join(test_data, cls)
    img = [im for im in os.listdir(cls_folder)]
    test_counts[cls] = len(img)
test_counts
```

```python
sample_img = cv2.imread(
    os.path.join(
        train_data,
        classes[0],
        os.listdir(os.path.join(train_data, classes[0]))[0]))
plt.imshow(sample_img)
```
